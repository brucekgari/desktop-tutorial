**Question 1**

How many different carriers are there in total in the database?

**Options:**

- 37

```python
%%sql

SELECT COUNT(DISTINCT(Description)) FROM carriers AS No_of_carriers;
```

**Question 2** 

How long was the longest delay before departure? 

**Options:**
- 1355 min
- 999 min
- 1099 min
- 588 min

```python
%%sql 

SELECT MAX(DepDelay) FROM flights AS longest_delay; 
```

**Question 3** 

How many flights departed on the 28th of January 2008?

**Options:**
- 540 908
- 20 149
- 19 495
- 18 056

```python
%%sql 

With CTE AS
(SELECT FlightNum, Date, Cancelled FROM Flights)
SELECT COUNT(FlightNum), Date
FROM CTE
WHERE Date = '2008/1/28'
    AND Cancelled = '0';
```

**Question 4**
 
 What is the distance between Midway Airport (MDW) and Houston Airport (HOU)?
 
 **Options:**
 - 611 km
 - 972 km
 - 1 121 km
 - 937 km

```python
%%sql

With CTE AS
(SELECT Origin, Dest, Distance FROM Flights)

SELECT * FROM CTE
WHERE Origin = 'MDW'
    AND Dest = 'HOU'
LIMIT 1;
```

**Question 5**

Which day of the week had the highest number of cancelled flights?(Note 1 = cancelled , 0 = not cancelled;)


**Options:**
- Monday(1)
- Wednesday(3) 
- Saturday(6) 
- Thursday (4) 

```python
%%sql 

With CTE AS
(SELECT DayOfWeek, Cancelled FROM Flights)

SELECT COUNT(DayOfWeek), DayOfWeek FROM CTE
WHERE Cancelled = '1'
GROUP BY DayOfWeek;
```

**Question 6**

How many airports have the word "International" in their name?

**Options:**

```python
%%sql

SELECT COUNT(airport) FROM airports
WHERE airport LIKE '%international%';
```

**Question 7**

What is the most produced model for the manufacturer "BOEING"?

**Options:**
- 717-200
- 737-7H4
- 757-222
- 737-3H4

```python
%%sql 

With CTE AS
(SELECT manufacturer, model FROM planes)

SELECT manufacturer, model, COUNT(model) FROM CTE 
WHERE manufacturer = 'BOEING'
GROUP BY model
ORDER BY COUNT(model) DESC;
```

**Question 8**

What manufacturer had the highest average delay time (DepDelay + ArrDelay)? 

**Options:**
- PAIR MIKE E
- AERONCA
- AVIAT AIRCRAFT INC
- BOEING OF CANADA LTD

```python
%%sql

SELECT manufacturer, DepTime, ArrDelay
FROM flights
INNER JOIN planes
ON flights.tailnum = planes.tailnum
```

**Question 9**

How many planes landed at Los Angeles International Airport? 

**Options:**
- 215 608
- 18 514
- 41 258
- 39 422

```python
%%sql 

SELECT airport, iata, COUNT(Dest)
FROM airports
INNER JOIN flights
ON airports.iata = flights.Dest
WHERE airport = 'Los Angeles International'
    AND Cancelled = '0';
```

**Question 10**

Which domestic carrier had the best on-time performance (OTP), where OTP is defined as the rate of on time flights with a 15min buffer on departure and arrival?

**Options:**
- American Airlines Inc
- Hawaiian Airlines Inc
- Comair Inc
- Aloha Airlines Inc

```python
%%sql 

SELECT Description, DepTime, ArrTime
FROM flights
INNER JOIN carriers
ON flights.UniqueCarrier = carriers.Code
WHERE ArrTime - DepTime = 15
```
